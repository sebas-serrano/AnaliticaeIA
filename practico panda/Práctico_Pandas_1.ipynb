```python
def is_organic(x):
  """Retorna si un pokemon es √≤rganico dado su typo 1"""
  organics = ['Grass','Bug','Normal','Poison','Ground']
  return x in organics
```

```python
pokemon =  get_my_data()
sns.set_theme(style="white")

# Generate a large random dataset

# Compute the correlation matrix


corr = pokemon.select_dtypes(exclude=['object']).drop(columns=['Total','#']).corr() ### TypeError: string dtypes are not allowed, use 'object' instead

# Mascara para solo graficar un tri√°ngulo
mask = np.triu(np.ones_like(corr, dtype=bool))


f, ax = plt.subplots(figsize=(11, 9))

# Generate un mapa de colores
cmap = sns.diverging_palette(100, 0, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask,
            cmap=cmap,
            vmax=.3,
            center=0,
            square=True,
            linewidths=.5,
            cbar_kws={"shrink": .5},
            annot=True
            )
plt.title("Mapa de correlaciones de atributos de Pokemons")
```
